```python
def fetch_top_coins_tickers(vs_currency='usd', limit=30):
    url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&ids&x_cg_demo_api_key=CG-RJK8hmewzGZMgR1PHncrkr84'
    params = {
        'vs_currency': vs_currency,
        'order': 'market_cap_desc',
        'per_page': limit,
        'page': 1,
        'sparkline': False,
        'price_change_percentage': '24h'
    }
    response = requests.get(url, params=params)
    if response.status_code != 200:
        print(f"Error fetching data: Status code {response.status_code}")
        return []

    data = response.json()

##Extract tickers
    tickers = [coin['name'].upper() for coin in data]
    return tickers
```

```python
import requests

def get_fear_greed_index():
    url = "https://api.alternative.me/fng/"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        # Extract the current value and sentiment classification
        result = data['data'][0]  # Most recent data point
        value = result['value']
        classification = result['value_classification']
        return value, classification
    else:
        return None, None

fear_greed_value, sentiment = get_fear_greed_index()
print(f"Fear & Greed Index: {fear_greed_value}, Sentiment: {sentiment}")
```

```python
url = 'https://api.coingecko.com/api/v3/search/trending'
headers = {
         'accept': 'application/json'
}
response = requests.get(url, headers = headers)
```

```python
if response.status_code == 200:
         # Convert the JSON response to a Python dictionary
         data = response.json()

         # Extract relevant fields from the JSON data (coins section)
         coin_data = []
         for coin in data.get('coins', []):
            coin_item = coin.get('item', {})
            coin_data.append({
            'id': coin_item.get('id', ''),
            'name': coin_item.get('name', ''),
            'symbol': coin_item.get('symbol', ''),
            'market_cap_rank': coin_item.get('market_cap_rank', 0),
            'score': coin_item.get('score', 0)
     })

         # Create a DataFrame from the extracted data
         df = pd.DataFrame(coin_data)

         # Display the DataFrame
         df

else:

         print(f"API request failed with status code: {response.status_code}")
```

```python
import requests
import datetime
import time
import pandas as pd
import ta

# Configuration
api_key = "CG-RJK8hmewzGZMgR1PHncrkr84"  # Replace with your actual CoinGecko Pro API key
coin_id = 'bitcoin'
vs_currency = 'usd'

# Calculate timestamps for one year range
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=365)
start_timestamp = int(time.mktime(start_date.timetuple()))
end_timestamp = int(time.mktime(end_date.timetuple()))

# API endpoint and parameters for CoinGecko API
url = f'https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart/range'
params = {
    'vs_currency': vs_currency,
    'from': start_timestamp,
    'to': end_timestamp,
    'x_cg_demo_api_key': 'CG-RJK8hmewzGZMgR1PHncrkr84'
}
headers = {
    'accept': 'application/json'
}

# Fetch data using requests with headers
response = requests.get(url, headers=headers, params=params)
if response.status_code != 200:
    raise Exception(f"Failed to fetch data: {response.status_code} - {response.text}")

data = response.json()

# Extract and process data as before
prices = data.get("prices", [])
market_caps = data.get("market_caps", [])
volumes = data.get("total_volumes", [])

df_prices = pd.DataFrame(prices, columns=["timestamp", "closing_price"])
df_market_caps = pd.DataFrame(market_caps, columns=["timestamp", "market_cap"])
df_volumes = pd.DataFrame(volumes, columns=["timestamp", "volume"])

df = df_prices.merge(df_market_caps, on="timestamp").merge(df_volumes, on="timestamp")
df['date'] = pd.to_datetime(df['timestamp'], unit='ms').dt.date
df['coin_name'] = coin_id.capitalize()
df = df[['date', 'coin_name', 'closing_price', 'volume', 'market_cap']]
df['closing_price'] = df['closing_price'].astype(float)
df['rsi_14'] = ta.momentum.rsi(df['closing_price'], window=14)
```
